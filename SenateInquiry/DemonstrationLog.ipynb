```r
# Start by reading some data using the inbuilt functions
data <- read.csv("https://cloudstor.aarnet.edu.au/plus/s/qgGU1srniF6yQsO/download")
```

```r
# Dimension of the data frame
dim(data)

# Structure of the dataframe
str(data)

# First few rows
head(data, 1)
```

```r
# Some basic data quality comparisons - later sessions might use the full 
# tidyverse, but here we're just going to touch on one little bit for some data 
# manipulation
install.packages("dplyr")
library(dplyr)
```

```r
# Check the basic submission statistics
data %>% count(Filetype)
# This is a bit less useful (imagine thousands of submissions)
data %>% count(Submission)
# Can address by 'arranging' by the count
data %>% count(Submission, name = 'file_count') %>% arrange(desc(file_count))

# Or we can go one step further and create a histogram of counts
data %>% count(Submission, name = 'file_count') %>% count(file_count, name = 'submission_count')
```

```r
submissions <- data %>% filter(Filetype == "Submission")
```

```r
# Install a package
install.packages("quanteda")
```

```r
# Load the quanteda library
library(quanteda)
```

```r
# Create a corpus from the dataframe, but we want to filter out the attachments - start to do some reshaping based on that.
corp <- corpus(
  submissions,
  docid_field = "Submission",
  text_field = "Text"
)

tokenised_corp = tokens(corp)

# Notice that by default there are non word characters here, we'll remove those later
print(tokenised_corp)
```

```r
quanteda::kwic(tokenised_corp, pattern = "TAFE", window = 5)

quanteda::kwic(tokenised_corp, pattern = phrase("regional inequality"), window = 5)
```
